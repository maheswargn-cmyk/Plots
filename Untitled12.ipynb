```python
import matplotlib.pyplot as plt

# --------------------------------------------------
# Survey data
# (Survey name, Effective wavelength in microns, Limiting magnitude)
# Limiting magnitudes are 5 point-source, typical stacked depths
# --------------------------------------------------
surveys = [
    ("LSST (r, coadd)", 0.62, 27.5),
    ("HSC-SSP (i)", 0.77, 26.8),
    ("DES (i)", 0.77, 24.1),
    ("Pan-STARRS (r)", 0.62, 23.3),
    ("ZTF (r)", 0.62, 21.0),
    ("OGLE-IV (I)", 0.80, 21.0),
    ("ASAS-SN (V)", 0.55, 17.0),
    ("UHS (J)", 1.25, 19.6),
    ("VISTA VHS (Ks)", 2.15, 20.0),
    ("VISTA VIKING (Ks)", 2.15, 21.5),
    ("UKIDSS LAS (K)", 2.20, 18.4),
    ("UKIDSS DXS (K)", 2.20, 21.0),
]
surveys1 = [
    ("VISTA VVV (Ks)", 2.15, 18.3),
    ("CFHT WIRCam (Ks)", 2.15, 21.0),
]

# Separate columns
names = [s[0] for s in surveys]
wavelengths = [s[1] for s in surveys]
magnitudes = [s[2] for s in surveys]
names1 = [s1[0] for s1 in surveys1]
wavelengths1 = [s1[1] for s1 in surveys1]
magnitudes1 = [s1[2] for s1 in surveys1]
# --------------------------------------------------
# Plot
# --------------------------------------------------
plt.figure(figsize=(9, 6))
plt.scatter(wavelengths, magnitudes)
plt.scatter(wavelengths1, magnitudes1)

# Add labels to each point
for name, x, y in zip(names, wavelengths, magnitudes):
    plt.text(x, y, name, fontsize=9, ha="left", va="bottom")
for name1, x1, y1 in zip(names1, wavelengths1, magnitudes1):
    plt.text(x1-.01, y1, name1, fontsize=9, ha="right", va="bottom")

plt.xlim(0.5, 2.6)
plt.xlabel("Wavelength (m)")
plt.ylabel("Limiting Magnitude (5, point source)")
plt.title("Limiting Magnitude vs Wavelength\nGround-based Optical & Infrared Imaging Surveys")

# Astronomical convention: fainter = larger magnitude
plt.gca().invert_yaxis()

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# -------------------------------------------------
# Imaging surveys (representative band & depth)
# (Name, wavelength [micron], limiting magnitude)
# -------------------------------------------------
imaging = [
    ("LSST (r, coadd)", 0.62, 27.5),
    ("HSC-SSP (i)", 0.77, 26.8),
    ("DES (i)", 0.77, 24.1),
    ("Pan-STARRS (r)", 0.62, 23.3),
    ("ZTF (r)", 0.62, 21.0),
    ("OGLE-IV (I)", 0.80, 21.0),
    ("ASAS-SN (V)", 0.55, 17.0),
    ("UHS (J)", 1.25, 19.6),
    ("VISTA VHS (Ks)", 2.15, 20.0),
    ("VISTA VIKING (Ks)", 2.15, 21.5),
    ("UKIDSS LAS (K)", 2.20, 18.4),
    ("UKIDSS DXS (K)", 2.20, 21.0),

]

# -------------------------------------------------
# Extragalactic spectroscopic surveys
# -------------------------------------------------
spec_extragal = [
    ("DESI", 0.62, 23.4),
    ("4MOST (LRS)", 0.62, 22.5),
    ("WEAVE (LR)", 0.62, 22.3),
    ("HETDEX", 0.48, 24.5),
    ("MOONS (galaxies)", 1.65, 22.5),
    ("Subaru PFS", 1.25, 23.0),
]

# -------------------------------------------------
# Galactic spectroscopic surveys
# -------------------------------------------------
spec_galactic = [
    ("SDSS-V (stars)", 0.55, 21.0),
    ("LAMOST", 0.62, 18.5),
    ("4MOST (HRS)", 0.62, 18.0),
    ("WEAVE (HR)", 0.62, 19.0),
    ("APOGEE-2", 1.65, 13.5),
]

# -------------------------------------------------
# Plot
# -------------------------------------------------
plt.figure(figsize=(10, 6))

# Imaging
plt.scatter(
    [x[1] for x in imaging],
    [x[2] for x in imaging],
    marker="o",
    label="Imaging surveys",
)

# Extragalactic spectroscopy
plt.scatter(
    [x[1] for x in spec_extragal],
    [x[2] for x in spec_extragal],
    marker="s",
    label="Spectroscopy (Extragalactic)",
)

# Galactic spectroscopy
plt.scatter(
    [x[1] for x in spec_galactic],
    [x[2] for x in spec_galactic],
    marker="^",
    label="Spectroscopy (Galactic)",
)

# -------------------------------------------------
# Annotate points
# -------------------------------------------------
for group in [imaging, spec_extragal, spec_galactic]:
    for name, x, y in group:
        plt.text(x, y, name, fontsize=8, ha="left", va="bottom")

# -------------------------------------------------
# Survey depth gap (imaging vs spectroscopy)
# -------------------------------------------------
plt.axhspan(
    24.5, 28,
    alpha=0.2,
    label="ImagingSpectroscopy Depth Gap"
)
# -------------------------------------------------
# Formatting
# -------------------------------------------------
plt.xlim(0.3,2.5)
plt.ylim(13,28)
plt.xlabel("Wavelength (m)")
plt.ylabel("Limiting Magnitude (typical survey depth)")
plt.title(
    "Imaging vs Spectroscopic Survey Depths\n"
)

# Astronomical magnitude convention
plt.gca().invert_yaxis()

plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
wavelength = np.linspace(0.3, 2.5, 300)
# =================================================
# Imaging surveys
# =================================================
imaging = [
    ("LSST (r, coadd)", 0.62, 27.5),
    ("HSC-SSP (i)", 0.77, 26.8),
    ("DES (i)", 0.77, 24.1),
    ("Pan-STARRS (r)", 0.62, 23.3),
    ("ZTF (r)", 0.62, 21.0),
    ("OGLE-IV (I)", 0.80, 21.0),
    ("ASAS-SN (V)", 0.55, 17.0),
    ("UHS (J)", 1.25, 19.6),
    ("VISTA VHS (Ks)", 2.15, 20.0),
    ("VISTA VIKING (Ks)", 2.15, 21.5),
    ("UKIDSS LAS (K)", 2.20, 18.4),
    ("UKIDSS DXS (K)", 2.20, 21.0),
]

# =================================================
# Spectroscopic surveys  Extragalactic
# =================================================
spec_extragal = [
    ("DESI", 0.62, 23.4),
    ("4MOST (LRS)", 0.62, 22.5),
    ("WEAVE (LR)", 0.62, 22.3),
    ("HETDEX", 0.48, 24.5),
    ("MOONS", 1.65, 22.5),
    ("Subaru PFS", 1.25, 23.0),
]

# =================================================
# Spectroscopic surveys  Galactic
# =================================================
spec_galactic = [
    ("SDSS-V", 0.55, 21.0),
    ("LAMOST", 0.62, 18.5),
    ("4MOST (HRS)", 0.62, 18.0),
    ("WEAVE (HR)", 0.62, 19.0),
    ("APOGEE-2", 1.65, 13.5),
]

# =================================================
# Large telescopes  spectroscopy limits (extragal.)
# =================================================
large_telescopes = [
    ("Keck (10 m)", 0.62, 26.0),
    ("GMT (24.5 m)", 0.62, 28.0),
    ("TMT (30 m)", 0.62, 29.0),
    ("ELT (39 m)", 0.62, 30.0),
]

# =================================================
# Plot
# =================================================
plt.figure(figsize=(11, 8))

# Imaging
plt.scatter(
    [x[1] for x in imaging],
    [x[2] for x in imaging],
    marker="o",
    s=80,
    label="Imaging surveys",
)

# Extragalactic spectroscopy
plt.scatter(
    [x[1] for x in spec_extragal],
    [x[2] for x in spec_extragal],
    marker="s",
    s=80,
    label="Spectroscopy (Extragalactic)",
)

# Galactic spectroscopy
plt.scatter(
    [x[1] for x in spec_galactic],
    [x[2] for x in spec_galactic],
    marker="^",
    s=80,
    label="Spectroscopy (Galactic)",
)

# Large telescopes
#plt.scatter(
#    [x[1] for x in large_telescopes],
#    [x[2] for x in large_telescopes],
#    marker="*",
#    s=200,
#    label="Large telescopes (spectroscopy)",
#)
# 8.2 m class (VLT / Subaru)
tel_8m = np.interp(
    wavelength,
    [0.3, 0.6, 1.0, 2.2, 2.5],
    [24.5, 25.5, 25.0, 23.0, 22.5],
)

# Keck / GTC (10 m class)
keck_gtc = np.interp(
    wavelength,
    [0.3, 0.6, 1.0, 2.2, 2.5],
    [25.0, 26.0, 25.5, 23.5, 23.0],
)

# 12.42 m telescope (scaled between 10 m and 13.7 m)
telescope_12m = np.interp(
    wavelength,
    [0.3, 0.6, 1.0, 2.2, 2.5],
    [25.6, 26.6, 26.1, 24.2, 23.7],
)

# 14.34 m telescope
tel_14m3 = np.interp(
    wavelength,
    [0.3, 0.6, 1.0, 2.2, 2.5],
    [26.2, 27.2, 26.7, 25.0, 24.5],
)

# GMT (24.5 m)
gmt = np.interp(
    wavelength,
    [0.3, 0.6, 1.0, 2.2, 2.5],
    [27.0, 28.0, 27.5, 26.0, 25.5],
)

# TMT (30 m)
tmt = np.interp(
    wavelength,
    [0.3, 0.6, 1.0, 2.2, 2.5],
    [28.0, 29.0, 28.5, 27.0, 26.5],
)

# ELT / E-ELT (39 m)
elt = np.interp(
    wavelength,
    [0.3, 0.6, 1.0, 2.2, 2.5],
    [29.0, 30.0, 29.5, 28.0, 27.5],
)

# -------------------------------------------------
# Plot
# -------------------------------------------------
plt.plot(wavelength, tel_8m, label="8.2 m (VLT / Subaru)")
plt.plot(wavelength, keck_gtc, label="Keck / GTC (10 m)")
plt.plot(wavelength, nlot, linestyle="--", linewidth=2.5, label="NLOT (13.7 m)")
#plt.plot(wavelength, tel_14m3, label="14.34 m (VLT / Subaru)")
plt.plot(wavelength, gmt, label="GMT (24.5 m)")
plt.plot(wavelength, tmt, label="TMT (30 m)")
plt.plot(wavelength, elt, label="ELT (39 m)")

# =================================================
# Labels
# =================================================
for group in [imaging, spec_extragal, spec_galactic, large_telescopes]:
    for name, x, y in group:
        plt.text(x, y, name, fontsize=8, ha="left", va="bottom")
plt.text(0.62-0.02, 26, "GTC", fontsize=8, ha="right", va="bottom")
# -------------------------------------------------
# Formatting
# -------------------------------------------------
plt.xlim(0.3,2.5)
plt.ylim(13,31)
plt.xlabel("Wavelength (m)")
plt.ylabel("Limiting Magnitude (typical survey depth)")
plt.title(
    "Imaging and Spectroscopic Survey Depths\n"
    "Including Large-Aperture Telescope Capabilities"
)

plt.gca().invert_yaxis()  # Astronomical magnitude scale
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
